```python
list_methods = []
for method in dir(list):
  if method.startswith("__"):
    continue
  list_methods.append(method)
print(list_methods)

set_methods = []
for method in dir(set):
  if method.startswith("__"):
    continue
  set_methods.append(method)
print(set_methods)

string_methods = []
for method in dir(str):
  if method.startswith("__"):
    continue
  string_methods.append(method)
print(string_methods)

tuple_methods = []
for method in dir(tuple):
  if method.startswith("__"):
    continue
  tuple_methods.append(method)
print(tuple_methods)

dict_methods = []
for method in dir(dict):
  if method.startswith("__"):
    continue
  dict_methods.append(method)
print(dict_methods)

basliklar = ["List Methods", "Set Methods", "String Methods", "Tuple Methods", "Dict Methods"]
classes = [list_methods, set_methods, string_methods, tuple_methods, dict_methods]

max_len = 0
for class_ in classes:
  if len(class_) > max_len:
    max_len = len(class_)

with open ("methods.txt", "w") as file:
  print("""
        """)
  for baslik in basliklar:
    print(baslik, end="")
    print(" " * (30 - len(baslik)), end="")
    file.write(baslik)
    file.write(" " * (30 - len(baslik)))
  print("""
        """)

  for i in range(max_len):
    file.write("\n")
    for class_ in classes:
      if i >= len(class_):
        print("-------", end="")
        print(" " * 23, end="")
        file.write("-------")
        file.write(" " * 23)
      else:
        print(class_[i], end="")
        print(" " * (30 - len(class_[i])), end="")
        file.write(class_[i])
        file.write(" " * (30 - len(class_[i])))
    print()
```
